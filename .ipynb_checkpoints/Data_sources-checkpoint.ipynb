```python
#Tried using crypto_tickers list but some currencies don't have data which brakes the code.
#we can add an exception but I need to make why 

since = binance.parse8601('2019-01-01T00:00:00z')
cols=['time','open','high','low','close','volume','symbol']
symbols=crypto_tickers#['BTC/USDT','ETC/BTC']
timeframe='1d'

cryptos_df= []
for i in symbols:
    try:
        price_batch = pd.DataFrame(binance.fetchOHLCV(i,timeframe, since, params = {}))
        price_batch[6]=i
        price_batch[0] = pd.to_datetime(price_batch[0], unit="ms")
        cryptos_df.append(price_batch)
    except Exception:
        print(i)#prints coins that did not 
cryptos_df2 = pd.concat(cryptos_df)
cryptos_df2.columns=cols
```
